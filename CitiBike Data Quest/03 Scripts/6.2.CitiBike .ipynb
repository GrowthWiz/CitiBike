```python
# Create a correlation matrix using pandas
#Select only the numeric columns and name it as numeric_df
numeric_df = df.select_dtypes(include=[np.number])
```

```python
# Convert the relevant columns to datetime if needed
df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time'] = pd.to_datetime(df['end_time'])
```

### Categorical Plots:

```python
# Use a histogram to visualize the distribution of the variables. 
# This way, you can determine sensible categories for the price ranges. 
# You don't want to end up with too few observations in any one of the categories.
# The argument "kde" add a line that encompasses the distribution

sns.histplot(numeric_df['trip_duration'], bins = 100, kde = True)

#The histogram reveals a prominent peak in frequency occurring at approximately 500 seconds!
```

```python
# Making trip category 8< minutes is "Short trip", 8-14 minutes is moderate and more than 14 minutes is long trip
df.loc[df['trip_duration'] < 480, 'trip_category'] = 'Short trip'
```

```python
df.loc[(df['trip_duration'] >= 480) & (df['trip_duration'] < 900), 'trip_category'] = 'Moderate trip'
```

```python
df.loc[df['trip_duration'] >= 900, 'trip_category'] = 'Long trip'
```

```python
df['trip_category'].value_counts(dropna = False)
```

```python
df.columns
```

```python
# Create a categorical plot in seaborn using the trip categories created above

sns.set(style="ticks")
g = sns.catplot(x="start_hour", y="trip_duration", hue="trip_category", data=df)

#There is lower density in long trips between midnight and 6 a.m.
```
